# <font color='Darkest Alice'>What is cricket

---

### Cricket Basics:

* Teams: Two teams play cricket.
* Players: Each team has 11 players.
* Objective: Score more runs than the opposing team.

### Game Structure:

* Innings: Teams take turns batting and bowling.
* Batting: Score runs by hitting the ball and running between wickets.
* Bowling: Try to get the batsmen out and limit their runs.

### Scoring:

* Runs: Batsmen score runs by hitting the ball and running.
* Boundaries: If the ball reaches the boundary, it's 4 runs. If it goes over without touching the ground, it's 6 runs.

### Dismissals:

* Bowled: The bowler knocks down the stumps.
* Caught: A fielder catches the ball.
* LBW (Leg Before Wicket): The ball hits the batsman's leg in front of the stumps.
* Run Out: A batsman is out of the crease when the bails are removed.
* Stumped: The wicketkeeper removes the bails while the batsman is out of the crease.

### Overs and Innings:

* Overs: A set of six deliveries by a bowler.
* Innings: Consists of one team batting and the other bowling.

### Extras:

* No Ball: Illegal delivery, gives the batting team a free run.
* Wide: Ball is too far for the batsman to hit, gives the batting team a free run.

### Winning:

* Runs: Team with more runs at the end of the innings wins.
* Wickets: If a team gets all the opposing players out, they win.
